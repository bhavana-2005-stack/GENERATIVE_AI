#Task 2: Fixed Size Chunking

```python
def fixed_chunking(text, chunk_size):
    chunks = []

    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i+chunk_size])

    return chunks
```

```python
chunks_100 = fixed_chunking(text,100)

for i,chunk in enumerate(chunks_100):
    print(f"\nChunk {i+1}")
    print(chunk)
    print("Length:",len(chunk))
```

#Task 8: Recommendation Report Table

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def retrieve(query, chunks):

    vectorizer = TfidfVectorizer()

    vectors = vectorizer.fit_transform(chunks + [query])

    similarity = cosine_similarity(
        vectors[-1],
        vectors[:-1]
    )

    idx = similarity.argmax()

    return chunks[idx]
```

```python
recommendations = pd.DataFrame({
    "Assistant Type":[
        "HR Policy Assistant",
        "Legal Document Assistant",
        "Medical Document Assistant",
        "Research Paper Assistant"
    ],
    "Recommended Chunking":[
        "Semantic",
        "Recursive",
        "Semantic",
        "Recursive"
    ],
    "Reason":[
        "Policies are naturally divided into sections",
        "Legal clauses span multiple paragraphs",
        "Medical topics should remain grouped",
        "Research content flows across sections"
    ]
})

recommendations
```
